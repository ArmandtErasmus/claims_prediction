```python
# libraries
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# data
df = pd.read_csv(r"data/historical_data.csv")
df.head()
```

```python
df["totalclaims"] = df["pastclaims"] + df["claims"]
df = df[["carcolour", "totalclaims"]]
```

```python
df2 = df.copy()
```

```python
total_white_cars = df2[df2["carcolour"] == "White"].shape[0]
total_blue_cars = df2[df2["carcolour"] == "Blue"].shape[0]
total_green_cars = df2[df2["carcolour"] == "Green"].shape[0]
total_black_cars = df2[df2["carcolour"] == "Black"].shape[0]
total_yellow_cars = df2[df2["carcolour"] == "Yellow"].shape[0]
total_red_cars = df2[df2["carcolour"] == "Red"].shape[0]
total_silver_cars = df2[df2["carcolour"] == "Silver"].shape[0]
```

```python
total_white_claims = df2.loc[df2["carcolour"] == "White", "totalclaims"].sum()
total_blue_claims = df2.loc[df2["carcolour"] == "Blue", "totalclaims"].sum()
total_green_claims = df2.loc[df2["carcolour"] == "Green", "totalclaims"].sum()
total_black_claims = df2.loc[df2["carcolour"] == "Black", "totalclaims"].sum()
total_yellow_claims = df2.loc[df2["carcolour"] == "Yellow", "totalclaims"].sum()
total_red_claims = df2.loc[df2["carcolour"] == "Red", "totalclaims"].sum()
total_silver_claims = df2.loc[df2["carcolour"] == "Silver", "totalclaims"].sum()
```

```python
white_rate = total_white_claims / total_white_cars
blue_rate = total_blue_claims / total_blue_cars
green_rate = total_green_claims / total_green_cars
black_rate = total_black_claims / total_black_cars
yellow_rate = total_yellow_claims / total_yellow_cars
red_rate = total_red_claims / total_red_cars
silver_rate = total_silver_claims / total_silver_cars
```

```python
car_counts = {
    "White": total_white_cars,
    "Blue": total_blue_cars,
    "Green": total_green_cars,
    "Black": total_black_cars,
    "Yellow": total_yellow_cars,
    "Red": total_red_cars,
    "Silver": total_silver_cars
}

claims = {
    "White": total_white_claims,
    "Blue": total_blue_claims,
    "Green": total_green_claims,
    "Black": total_black_claims,
    "Yellow": total_yellow_claims,
    "Red": total_red_claims,
    "Silver": total_silver_claims
}

df_claims = pd.DataFrame({
    "carcolour" : list(claims.keys()),
    "claims" : list(claims.values()),
    "num_cars": list(car_counts.values())
})
df_claims.head()
```

```python
rates = {
    "White": white_rate,
    "Blue": blue_rate,
    "Green": green_rate,
    "Black": black_rate,
    "Yellow": yellow_rate,
    "Red": red_rate,
    "Silver": silver_rate
}

df_rates = pd.DataFrame({
    "carcolour": list(rates.keys()),
    "claim_rate": list(rates.values())
})
df_rates.head()
df_rates["claim_rate"] = df_rates["claim_rate"] * 1000
```

```python
fig = px.bar(
    df_rates,
    x="carcolour",
    y="claim_rate",
    title="Claims per 1,000 Cars",
    labels={"carcolour": "Car Colour", "claim_rate": "Claims per 1,000 Cars"},
    color_discrete_sequence=["#ffc247"],
    text = "claim_rate"
)

fig.update_traces(texttemplate='%{text:.0f}', textposition='inside')
fig.update_layout(xaxis_tickangle=0)

fig.show()
```

```python
df2["one_claim"] = (df2["totalclaims"] == 1).astype(int)
df2["multi_claim"] = (df2["totalclaims"] > 1).astype(int)

df_repeat = df2.groupby("carcolour").agg(
    total_cars=("carcolour", "count"),
    one_claim_count=("one_claim", "sum"),
    multi_claim_count=("multi_claim", "sum")
).reset_index()

df_repeat["one_claim_rate"] = df_repeat["one_claim_count"] / df_repeat["total_cars"] * 1000
df_repeat["multi_claim_rate"] = df_repeat["multi_claim_count"] / df_repeat["total_cars"] * 1000

```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=df_repeat["carcolour"],
    y=df_repeat["one_claim_rate"],
    name="1 Claim per 1,000 Cars"
))

fig.add_trace(go.Bar(
    x=df_repeat["carcolour"],
    y=df_repeat["multi_claim_rate"],
    name=">1 Claim per 1,000 Cars"
))

fig.update_layout(
    title="Claim Frequency by Car Colour (One vs Multiple Claims)",
    xaxis_title="Car Colour",
    yaxis_title="Claims per 1,000 Cars",
    barmode="group",
    showlegend=True,
    yaxis=dict(range=[0,500]),
    template="plotly_white",
    height=500,
    margin=dict(t=40)
)

fig.show()
```

 Exposure Bias

Exposure = how many units exist in each group.

If one category has more cars, it will naturally have more total claims, even if the cars are not riskier.

Example from your data:

White cars = many

Green cars = much fewer

So white cars will show:

more total claims 

but not necessarily higher claim rates 

Exposure bias = confusing total counts with underlying risk.

Your multi-claim and one-claim plots removed exposure bias by using claims per 1,000 cars.

 Base-Rate Fallacy

The base-rate fallacy happens when you ignore the base rate (the underlying proportions) and draw conclusions from raw totals.

In your case:

White cars have the most claims, therefore white cars are more risky.

This ignores the base rate:

There are simply more white cars, so of course they have more claims.

The fallacy:
Assuming the higher number means higher risk, without correcting for how many cars are in each colour group.

Once you adjust for the base rate (by dividing by total cars), the riskiness difference disappears.

 Putting them together in your context
Base-rate fallacy

 You misinterpret total claims as risk because you arent accounting for the number of cars in each colour.

Exposure bias

 The dataset is uneven: white cars dominate the exposure, so they dominate the total claims.

Together, they cause misleading conclusions like:

White cars claim more  white cars are riskier.

But your frequency plots showed:

one-claim rate  constant

multi-claim rate  constant

total claims differences  risk differences

Meaning:
Car colour is not a predictive risk factor once exposure is controlled for.

```python
# Create a dictionary of claim rates
claim_rates = {
    "White": white_rate,
    "Blue": blue_rate,
    "Green": green_rate,
    "Black": black_rate,
    "Yellow": yellow_rate,
    "Red": red_rate,
    "Silver": silver_rate
}

# Compute relative risk vs White
white = claim_rates["White"]
relative_risk = {colour: rate / white for colour, rate in claim_rates.items()}
```

```python
# Build dataframe for convenience
df_rel = pd.DataFrame({
    "carcolour": ["White", "Blue", "Green", "Black", "Yellow", "Red", "Silver"],
    "rate": [white_rate, blue_rate, green_rate, black_rate, yellow_rate, red_rate, silver_rate]
})

# Set white as baseline
baseline = df_rel.loc[df_rel["carcolour"] == "White", "rate"].iloc[0]

# Relative risk vs white
df_rel["relative_risk"] = df_rel["rate"] / baseline

# Percentage difference
df_rel["pct_diff"] = (df_rel["relative_risk"] - 1) * 100

# Plot
fig = go.Figure()

fig.add_trace(go.Bar(
    x=df_rel["carcolour"],
    y=df_rel["relative_risk"],
    text=df_rel["pct_diff"].apply(lambda x: f"{x:+.2f}%"),  # e.g. +2.20%
    textposition="inside",
    marker_color="#ff9f1a",
    name="Relative Risk vs White"
))

# Baseline line at 1.0
fig.add_hline(y=1.0, line_dash="dash", line_color="black")

fig.update_layout(
    title={
        "text" : "Relative Claim Frequencies by Vehicle Colour Using White as the Reference Category",
        "x" : 0.5,
        "xanchor": "center"},
    yaxis_title="Relative Claim Frequency (White = 1.00)",
    xaxis_title="Vehicle Colour",
    uniformtext_minsize=10,
    uniformtext_mode="hide",
    yaxis=dict(range=[min(relative_risk.values()) - 0.05,
                      max(relative_risk.values()) + 0.05]),
    template="plotly_white",
    height=500,
    margin=dict(t=40),
)

fig.show()
```

```python
# Relative risk of vehicle claims by colour, using White as the reference category (relative risk = 1.00). Bars show the claim frequency for each vehicle colour relative to White, and the text above each bar indicates the percentage difference compared to White. A dashed line at 1.0 indicates the baseline frequency.
```

In this case, exposure simply means:

 How many cars of each colour are in the dataset

(or in actuarial terms: how many units at risk).

 Why is this called exposure?

In insurance, exposure refers to the amount of risk the insurer is exposed to.

For example:

If you insure 26,910 white cars,

And only 4,692 yellow cars,

then you are exposed to far more potential claims from white cars.

So naturally:

White cars will produce more total claims,

Even if they have the exact same claim rate as yellow cars.

This is why total claims alone are misleading.

 Put simply:
Exposure = number of insured items that could generate a claim.
